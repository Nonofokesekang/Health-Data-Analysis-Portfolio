```python
# HEALTH DATA EXPLORATION PROJECT
# Author: Nonofo Kesekang

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("../data/sample_health_dataset.csv")

# Preview data
df.head()

# Basic info
df.info()

# Summary statistics
df.describe()

# Check missing values
df.isnull().sum()

# Simple cleaning: fill missing ages
df['Age'] = df['Age'].fillna(df['Age'].median())

# Gender distribution
df['Gender'].value_counts().plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# Age distribution
df['Age'].plot(kind='hist', bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()

# Diagnosis counts
df['Diagnosis'].value_counts().plot(kind='bar')
plt.title("Common Diagnoses")
plt.xlabel("Diagnosis")
plt.ylabel("Frequency")
plt.show()

# Save cleaned dataset
df.to_csv("../data/sample_health_dataset_cleaned.csv", index=False)
```
